```python
import pandas as pd
```

1. Importing Pandas and Data

```python
data = pd.read_csv('/content/drive/MyDrive/Data Analysis Portfolio - Saniya Shaikh/Machine Learning Projects - Python/Python Data File/winequality-red.csv')
data.describe()
data.shape
data.info()
data.tail()
data.head()
```

2. Working with Columns

```python
data.columns # list all column names
data['pH'] #Access a single column
data[['alcohol','quality']] #Access multiple columns
```

3. Selecting Rows & Filtering Data

```python
data.iloc[5] # Row with index label 3

#Filter rows
data[data['total sulfur dioxide'] > 50]
```

4. Creating & Modifying Columns

```python
data['New Column'] = data['pH'] + data['alcohol']
data.head()
```

5. Handling Missing Data

```python
data.isnull().sum() #checking for null values
data.dropna() #drop rows with missing values
data.fillna(0) #fill missing values with 0
```

6. Sorting and Resetting Index

```python
data.sort_values(by = 'total sulfur dioxide' , ascending= True) # Sorting data in ascending order
```

7. Grouping & Aggregation

```python
data.groupby('quality').mean() # Grouping and aggregating data
```

8. Merging, Joining, Concatenating

```python
pd.concat([data,data]) # Concatenating data
```
